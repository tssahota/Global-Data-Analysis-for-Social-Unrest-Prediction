```python
#### World Map of Events with GoldsteinScale > 5 and avgTone < 0###
df = pd.read_csv('/content/drive/MyDrive/GDELT_Full/GDELT_master.csv')
```

```python
df1 = df[['SQLDATE','MonthYear','Year','Actor1Code','Actor1CountryCode','ActionGeo_CountryCode','ActionGeo_Lat','ActionGeo_Long','IsRootEvent','EventRootCode','GoldsteinScale','AvgTone','NumArticles']]
```

```python
df2 = df1[(df1['GoldsteinScale'] <=-5) & (df1['AvgTone'] <=-5) & (df1['NumArticles'] > 50)]
#df2 = df2.groupby('MonthYear').agg({'GoldsteinScale': ['count','mean'],'AvgTone': ['count','mean']})
df2.head()
#df2['GoldsteinScale'] = (df2['GoldsteinScale']*-1) - 4
tone_max= df2['GoldsteinScale'].max()
tone_min= df2['GoldsteinScale'].min()
#df2['GoldsteinScale'] = (df2['GoldsteinScale'] - tone_min)/tone_max
df2.to_csv('./web_dev/apps/analysis_data/task1.csv')
```

```python
fig = px.scatter_geo(df2, lat=df2['ActionGeo_Lat'],lon=df2['ActionGeo_Long'],
                     hover_name="ActionGeo_CountryCode",  color='GoldsteinScale',
                     opacity=0.70,color_continuous_scale=px.colors.sequential.OrRd[::-1],
                     animation_frame="Year", #size = df2['GoldsteinScale'],
                     projection="natural earth", width=1000, height=700
                    )

fig.show()
```

```python
df3 = df1[df1['NumArticles'] >= 0]
df3.head()
df3 = df3.groupby(['Year','Actor1CountryCode']).mean().reset_index()
#df3.to_csv('./web_dev/apps/analysis_data/live.csv')
```

```python
df3=df3[df3['SQLDATE']>20210301]
std_avg = df3['GoldsteinScale'].mean()
std_max = df3['GoldsteinScale'].max()
df3['GoldsteinScale_std'] = df3['GoldsteinScale'] - std_avg

fig2 = px.choropleth(df3, locations="Actor1CountryCode",
                    color='GoldsteinScale_std', # lifeExp is a column of gapminder
                    hover_name="Actor1CountryCode", # column to add to hover information
                    locationmode="ISO-3",
                    hover_data = ['GoldsteinScale','AvgTone'],
                    color_continuous_scale=px.colors.diverging.RdBu)
fig2.show()
```
